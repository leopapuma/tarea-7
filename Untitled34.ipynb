```python
# usaremos esta libreria para resolver el problema pedido
import sympy as sp


# Definición el  problema que aparece en la imagen

x = sp.symbols('x')

# escribimos el polinomio de aproximacion
a0, a1, a2, a3 = sp.symbols('a0 a1 a2 a3')
uN = a0 + a1*x + a2*x**2 + a3*x**3

#escribimos la condciones como aparecen abajo
# u(0)=0  → a0=0
# u'(0)=2 → a1=2
a0_sol = 0
a1_sol = 2

uN = uN.subs({a0: a0_sol, a1: a1_sol})


# Epsilon = u'' + u - x

uN_prime = sp.diff(uN, x)
uN_pp = sp.diff(uN_prime, x)

epsilon = uN_pp + uN - x


# escribimos la Colocación en x = 1/2 y x = 1

eq1 = sp.simplify(epsilon.subs(x, sp.Rational(1,2)))
eq2 = sp.simplify(epsilon.subs(x, 1))

sol = sp.solve([eq1, eq2], [a2, a3])
sol
# con esta parte de codigo obtenderemos los valores de los coeficientes a2 y a3
```

```python

#  Construimos  la solución aproximada

uN_approx = uN.subs(sol)
sp.simplify(uN_approx)
```

```python

# escribimos la  Solución exacta

u_exact = sp.sin(x) + x


# esribimos el error que aparece de la sig manera  error: |u_exact - uN_approx|

error = sp.simplify(u_exact - uN_approx)
error
```

```python
for val in [0, 0.25, 0.5, 0.75, 1]:
    print(f"x={val}: error = {error.subs(x,val)}")
```

```python
#usaremos estas librerias para construir las graficas correspondientes
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt



# ponemos la Solución exacta
u_exact = sp.sin(x) + x

# escribimos el Error
error = sp.simplify(u_exact - uN_approx)


 #  construimos la Gráfica

xs = np.linspace(0, 1, 200)
uN_vals = [float(uN_approx.subs(x, val)) for val in xs]
u_exact_vals = [float(u_exact.subs(x, val)) for val in xs]
error_vals = [float(error.subs(x, val)) for val in xs]

plt.figure(figsize=(12,6))

# Gráfica de soluciones
plt.subplot(1,2,1)
plt.plot(xs, u_exact_vals, label='Solución exacta u(x)')
plt.plot(xs, uN_vals, '--', label='Aproximación u_N(x)')
plt.title("Comparación entre solución exacta y aproximada")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.legend()
plt.grid(True)

# Gráfica del error
plt.subplot(1,2,2)
plt.plot(xs, error_vals, color='red')
plt.title("Error |u(x) − u_N(x)|")
plt.xlabel("x")
plt.ylabel("error")
plt.grid(True)

plt.tight_layout()
plt.show()
#construimos con este codigo la grafica de la comparacion de solucion exacta y la aproximada y el error
```
